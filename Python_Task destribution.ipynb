```python
#Remove features which have a very low variance (less than 0.01)
tasks_var = []
for i in range(1,len(tasks.columns)):
    if tasks.iloc[:,i].var() < 0.01:
        tasks_var.append(i)
tasks = tasks.drop(tasks.columns[tasks_var],axis=1)
tasks.shape
```

1.3 Scale features to range [-1,1]

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

def scale_change(x):
    return 2*x-1
```

```python
#transform task features into -1 to 1 scale
tasks.iloc[:,1:] = scaler.fit_transform(tasks.iloc[:,1:])
tasks.iloc[:,1:] = tasks.iloc[:,1:].apply(scale_change)
tasks.head()
```

```python
#transform supplier features into -1 to 1 scale
suppliers.iloc[:,1:] = scaler.fit_transform(suppliers.iloc[:,1:])
suppliers.iloc[:,1:] = suppliers.iloc[:,1:].apply(scale_change)
suppliers.head()
```

1.4 absolute correlation of all pairs of features and visualize

```python
corr_matrix =tasks.corr().abs()
corr_matrix.style.background_gradient(cmap="GnBu")
```

```python
# Identify features with correlation of more than and equal to 0.8
def correlation(dataset, threshold):
    col_corr=set() #Set of all the names of correlated columns
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>=threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
#Drop the columns
corr_features=correlation(tasks,0.8)
tasks=tasks.drop(corr_features,axis=1)
tasks
```

2.2 Create one figure (using boxplots) that shows, for each supplier, the distribution of Errors (Eq. 1) 

```python
#create cost_pivot table
cost_pivot = pd.pivot(cost,index='Task ID', values='Cost', columns='Supplier ID')

#sort columns 
cost_pivot_columns_old = cost_pivot.columns.values.tolist()

cost_columns_old = np.array([])
for i in cost_pivot_columns_old:
    j = i.lstrip('S')
    j = int(j)
    cost_columns_old = np.append(cost_columns_old,[j])
cost_columns_sorted = np.sort(cost_columns_old)

cost_columns_new = np.array([])
for i in cost_columns_sorted:
    j = ('S'+str(int(i)))
    cost_columns_new = np.append(cost_columns_new,[j])

cost_pivot = cost_pivot.reindex(columns=cost_columns_new)
cost_pivot
```

```python
#create error_pivot table
T=np.array([])
S=np.array([])
E=np.array([])
for i,t_ID in enumerate(cost_pivot.index):
    for j,s_ID in enumerate(cost_pivot.columns):
        error = cost_pivot.iloc[i].min() - cost_pivot.iloc[i,j] 
        T = np.append(T,[t_ID])
        S = np.append(S,[s_ID])
        E = np.append(E,[error])

error_table = pd.DataFrame(data = {'Task ID':T,'Supplier ID':S,'Error':E})
error_table_pivot = pd.pivot(error_table,index='Task ID', values='Error', columns='Supplier ID')
error_table_pivot = error_table_pivot.reindex(columns=cost_columns_new)
display(error_table_pivot)
print('Mean of error values is',np.mean(np.mean(error_table_pivot)))
```

```python
#calculate RMSE score
RMSE_score = np.array([])
for i,s_ID in enumerate(error_table_pivot.columns):
    RMSE = (np.sum((error_table_pivot.iloc[:,i])**2) / len(error_table_pivot.index))**0.5
    RMSE = round(RMSE,5)
    RMSE_score = np.append(RMSE_score,[RMSE])
```

```python
plt.figure(figsize=(15, 10))
plt.boxplot(error_table_pivot,labels=cost_columns_new)
plt.title('Distribution of Error Values',fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
plt.xlabel('Supplier ID')
plt.ylabel('Error Value')
for i,score in enumerate(RMSE_score): 
    plt.text(i+0.3,error_table_pivot.iloc[:,i].median()+0.002, score, size='small', color='blue')
plt.text(17,0.005, '*Blue text is the RMSE of each supplier for all tasks(Not median!)', size='large', color='blue', weight='semibold')
plt.show()
print('Average RMSE score is',np.mean(RMSE_score))
```

# 3. ML model fitting and scoring

3.1 Data integration

```python
#combine the data
df = pd.merge(cost, tasks, how = 'inner', on = 'Task ID')
df = pd.merge(df, suppliers, how = 'inner', on = 'Supplier ID')
df = df.rename(columns={'Cost': 'Cost_old'})
df['Cost'] = df['Cost_old']
df = df.drop(['Supplier ID','Cost_old'],axis=1)
df.head()
```

```python
#split the dataset
X = df.drop(['Task ID','Cost'],axis=1)
y = df['Cost']
Groups=df['Task ID']
```

3.2 Split Train/Test data

```python
#TestGroup
TestGroup = pd.DataFrame(Groups.value_counts()).sample(n =20, random_state = 0, replace=False).drop('Task ID',axis=1)
TestGroup
```

```python
X_train = df[df['Task ID'].isin(TestGroup.index)==False].drop(['Task ID','Cost'],axis=1)
X_train
```

```python
y_train = df[df['Task ID'].isin(TestGroup.index)==False][['Cost']]
y_train
```

```python
X_test = df[df['Task ID'].isin(TestGroup.index)==True].drop(['Task ID','Cost'],axis=1)
X_test
```

```python
y_test = df[df['Task ID'].isin(TestGroup.index)==True][['Cost']]
y_test
```

3.3 Model fitting

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()

gbr.fit(X_train,y_train)
print(gbr.score(X_test,y_test))
```

3.4 Error/RMSE score

```python
# Predict the minimum cost for each task
##pick up Task ID of the test data as np.array
X_test_r = df[df['Task ID'].isin(TestGroup.index)==True].drop(['Cost'],axis=1)
X_test_id = X_test_r.groupby('Task ID').count()
Test_task_id = np.array(X_test_id.index.values.tolist())

##predict the cost for each task and pick up the minimum cost
pred_best = np.array([])
for i,task_id in enumerate (Test_task_id):
    X_test_rr = X_test_r.groupby('Task ID').get_group(task_id)
    best = gbr.predict(X_test_rr.iloc[:,1:]).min()
    pred_best = np.append(pred_best,[best])
print(pred_best)
    
##Create table for check the correspondence with task id
pred_best_table = pd.DataFrame(data=np.column_stack((Test_task_id,pred_best)),columns=['Test_task_id','predicted_minimum_cost'])
pred_best_table
```

```python
# Select the best suppliers
##Select the closest actual cost from the data to the predicted minimum cost
cost_pivot_r = cost_pivot[cost_pivot.index.isin(TestGroup.index)==True]
cost_pivot_array = cost_pivot_r.to_numpy()
pred_cost = np.array([])
pred_sup  = np.array([])
for i in range(len(pred_best)):
    idx = np.argmin(np.abs(cost_pivot_array[i]- pred_best[i]))
    pred_cost = np.append(pred_cost,[cost_pivot_array[i,idx]])
    pred_sup = np.append(pred_sup,[cost_columns_new[idx]])
print(pred_cost)
print(pred_sup)

##Create table for check the correspondence with task id
Pred_sup_table = pd.DataFrame(data=np.column_stack((Test_task_id,pred_sup,pred_cost)),columns=['Test_task_id','predicted_best_supplier','cost'])
Pred_sup_table
```

```python
# Calculate the Error values

##Actual minimum cost - the cost of the supplier chosen by ML model
error_value = np.array([])
best_cost = np.array([])
best_sup  = np.array([])
for i in range(len(pred_best)):
        best_cost = np.append(best_cost,[cost_pivot_array[i].min()])
        idx = np.argmin(cost_pivot_array[i])
        best_sup = np.append(best_sup,[cost_columns_new[idx]])
        error = cost_pivot_array[i].min()- pred_cost[i]
        error_value = np.append(error_value,[error])
print(error_value)

##Create table for check the correspondence with task id
Error_table = pd.DataFrame(data=np.column_stack((Test_task_id,error_value,best_sup,best_cost,pred_sup,pred_cost)),columns=['Test_task_id','Error Values','best_supplier','best_cost','predicted_best_supplier','predicted_best_cost'])
Error_table
```

```python
# Calculate the RMSE score
RMSE = (np.sum(error_value**2)/len(error_value))**0.5
print('Average Error Value is',np.mean(error_value))
print('RMSE score of this model is',RMSE)
```

# 4.  Cross-validation

```python
#Leave-One-Group-Out
from sklearn.model_selection import LeaveOneGroupOut,cross_val_score
logo = LeaveOneGroupOut()

print('The number of cross-validation is',logo.get_n_splits(groups=Groups))

y_min = np.array([])
for train_index, test_index in logo.split(X, y, groups=Groups):
    X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train, y_test = y[train_index], y[test_index]
    y_min = np.append(y_min,[y_test.min()])
```

```python
#Define a score function by using make_scorer
from sklearn.metrics import make_scorer
def error(y,y_pred):
    pred_min_cost = y_pred.min()
    y_array = y.to_numpy()
    idx = np.argmin(np.abs(y_array - pred_min_cost))
    pred_best_cost = y_array[idx]
    actual_min_cost = y.min()
    error_value = actual_min_cost - pred_best_cost
    return error_value
score = make_scorer(error)
```

```python
#Result
scores = cross_val_score(gbr, X, y, groups=Groups ,cv=logo, scoring = score, n_jobs=-1)
print('The scores returned by cross-validation:',np.mean(scores))
print('RMSE score:',(np.sum(scores**2)/len(scores))**0.5)
```

# 5. Hyper-parameter optimization

```python
from sklearn.model_selection import GridSearchCV
param_grid = dict(n_estimators = [300,500],learning_rate=[0.01,0.1],max_depth=[3,5])
gs = GridSearchCV(gbr, param_grid, cv=logo,scoring =score,n_jobs=-1)
```

```python
# Find the search
gs.fit(X, y, groups=Groups)
print("Best parameters found:", gs.best_params_)
print("Mean CV score of best parameters:", gs.best_score_)

#Modify models with best params
best_learning_rate = gs.best_params_['learning_rate']
best_max_depth = gs.best_params_['max_depth']
best_n_estimators = gs.best_params_['n_estimators']
best_gbr = GradientBoostingRegressor(learning_rate=best_learning_rate, max_depth=best_max_depth,n_estimators=best_n_estimators)

best_scores = cross_val_score(best_gbr, X, y, groups=Groups ,cv=logo, scoring = score, n_jobs=-1)
print('RMSE score of best parameters:',(np.sum(best_scores**2)/len(best_scores))**0.5)
```
